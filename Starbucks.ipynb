```python
import numpy as np
import pandas as pd
# 读取数据，将文件转换为DataFrame格式
data = pd.read_csv('directory.csv')
data.head()
```

```python
# 用于查看数值型数据的分布情况
data.describe()
```

```python
# 用于查看各字段的数据类型，以及缺失情况
data.info()
```

```python
# 查看唯一品牌的类型
data['Brand'].unique()
```

```python
# 只获取品牌为Starbucks
data = data[data['Brand'] == 'Starbucks']
data['Brand'].unique()
```

```python
# 各个字段缺失值的数量
data.isnull().sum()
```

```python
# 查看缺失城市是哪些
data[data['City'].isnull()]
```

```python
def fill_na(x):
    return x

# 用国家字段填充到城市字段上
data['City'] = data['City'].fillna(fill_na(data['State/Province']))
data[data['Country'] == 'EG']
```

```python
# data['Country'][data['Country'] == 'TW'] = 'CN'
# 发现台湾被美国当成一个国家，不能忍，把他重新赋值到中国来
data.loc[data['Country'][data['Country'] == 'TW']] = 'CN'
```

```python
country_count = data['Country'].value_counts()[0:10]
country_count
```

```python
import matplotlib.pyplot as plt
#import matplotlib
#matplotlib.matplotlib_fname()  # 将会获得matplotlib包所在文件夹
# 指定默认字体
plt.rcParams['font.sans-serif'] = ['simhei']

# 用来正常显示坐标轴上的负号（‘-’）
plt.rcParams['axes.unicode_minus'] = False

# %matplotlib具体作用是当你调用matplotlib.pyplot的绘图函数plot()进行绘图的时候，
# 或者生成一个figure画布的时候，可以直接在你的python console里面生成图像
# “%matplotlib inline”就是模仿命令行来访问magic函数的在IPython中独有的形式
%matplotlib inline

plt.title('全球星巴克数量前十的国家')
country_count.plot(kind = 'bar')
```
